```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
```

```python
df = pd.read_csv('crypto-markets.csv')
df.info()
```

```python
df.head(1)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['slug'] = df['slug'].astype('category')
df['symbol'] = df['symbol'].astype('category')
df['name'] = df['name'].astype('category')
df.dropna(axis=0, how='any', inplace=True)
df.info()
```

```python
mask = df['ranknow'] <= 100
mask = df[mask]
mask = mask.drop(['symbol', 'name', 'ranknow', 'high', 'low', 'open', 'volume', 'market', 'close_ratio', 'spread'], 1)
mask = mask.pivot(index='date', columns='slug', values='close')
mask.dropna(axis=0, how='all', inplace=True)
mask.head()
```
